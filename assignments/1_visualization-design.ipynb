```python
import pandas as pd

antibiotics = pd.read_csv("../data/antibiotics.csv")
```

```python
antibiotics.columns = antibiotics.columns.str.strip()
```

```python
antibiotics
```

```python
print(antibiotics["Penicilin"].min(),
antibiotics["Streptomycin"].min(),
antibiotics["Neomycin"].min())

print(antibiotics["Penicilin"].max(),
antibiotics["Streptomycin"].max(),
antibiotics["Neomycin"].max())
```

```python
import altair as alt

grid_values = [0.001, 0.01, 0.1, 1, 10, 100, 1000]

chart1 = alt.Chart(antibiotics).mark_bar().encode(
    x=alt.X("Penicilin:Q", scale=alt.Scale(type="log", clamp=True, domain=[0.001, 1000]), axis=alt.Axis(grid=True, values=grid_values)),
    y=alt.Y("Bacteria"),
    color="Gram Staining",
).properties(
    width=250,
    height=350 
)

chart2 = alt.Chart(antibiotics).mark_bar().encode(
    x=alt.X("Streptomycin:Q", scale=alt.Scale(type="log", clamp=True, domain=[0.001, 1000]), axis=alt.Axis(grid=True, values=grid_values)),
    y=alt.Y("Bacteria", axis=alt.Axis(labels=False, title=None)),
    color="Gram Staining",
).properties(
    width=250,
    height=350 
)

chart3 = alt.Chart(antibiotics).mark_bar().encode(
    x=alt.X("Neomycin:Q", scale=alt.Scale(type="log", clamp=True, domain=[0.001, 1000]), axis=alt.Axis(grid=True, values=grid_values)),
    y=alt.Y("Bacteria", axis=alt.Axis(labels=False, title=None)),
    color="Gram Staining",
).properties(
    width=250,
    height=350
)


alt.hconcat(chart1, chart2, chart3).properties(
    title=alt.TitleParams("The Effectiveness of Antibiotics on Bacteria", anchor="middle")
)
```
